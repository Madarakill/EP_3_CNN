```python
# Словарь перевода итальянских названий животных на английский
translate = {
    "cane": "dog", "cavallo": "horse", "elefante": "elephant",
    "farfalla": "butterfly", "gallina": "chicken", "gatto": "cat",
    "mucca": "cow", "pecora": "sheep", "scoiattolo": "squirrel", "ragno": "spider"
}
# Корневая директория датасета
dataset_root = "animals10/raw-img"

# Структура для хранения путей к изображениям
dataset_records = {
    "image_path": [],
    "classes": []
}

italian_labels = [
    k for k in translate.keys()
    if os.path.isdir(os.path.join(dataset_root, k))
]

for classes in italian_labels:
    class_dir = os.path.join(dataset_root, classes)
    image_files = []
    for ext in ("*.jpeg", "*.jpg", "*.png"):
        image_files.extend(glob.glob(os.path.join(class_dir, ext)))

    dataset_records["image_path"].extend(image_files)
    dataset_records["classes"].extend([classes] * len(image_files))

df_data = pd.DataFrame(dataset_records)
#вывод информации 
print(f"Найдено изображений: {len(df_data)}")
print("Классы ", sorted(df_data['classes'].unique()))
df_data.head()
```

```python
# Преобразования
train_transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5],
                         [0.5, 0.5, 0.5])
])

# Для тестовой и валидационной выборки
test_transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5],
                         [0.5, 0.5, 0.5])
])
base_dataset = ImageFolder(root=dataset_root)
```

```python
# Разделение на train, test, val
train_size = int(0.7 * len(base_dataset))
test_val_size = len(base_dataset) - train_size
train_data, test_val_data = random_split(base_dataset, [train_size, test_val_size])

test_size = int(0.5 * len(test_val_data))
val_size = len(test_val_data) - test_size
test_data, val_data = random_split(test_val_data, [test_size, val_size])

# Применяем разные преобразования
class TransformSubset(torch.utils.data.Dataset):
    def __init__(self, subset, transform=None):
        self.subset = subset
        self.transform = transform
        
    def __getitem__(self, index):
        x, y = self.subset[index]
        if self.transform:
            x = self.transform(x)
        return x, y
        
    def __len__(self):
        return len(self.subset)

# Создание датасетов
train_data = TransformSubset(train_data, transform=train_transform)
test_data = TransformSubset(test_data, transform=test_transform)
val_data = TransformSubset(val_data, transform=test_transform)

len(train_data), len(test_data), len(val_data)
```

```python
# Загрузчики
batch_size = 32
train_loader = DataLoader(train_data,
                          batch_size=batch_size,
                          shuffle=True,
                          num_workers=2)

test_loader = DataLoader(test_data,
                         batch_size=batch_size,
                         shuffle=False,
                         num_workers=2)

val_loader = DataLoader(val_data,
                       batch_size=batch_size,
                       shuffle=False,
                       num_workers=2)
```

```python
italian_labels = sorted(df_data['classes'].unique())
translated_labels = [translate[it] for it in italian_labels]

# Сопоставление индекса
idx_to_italian = {i: label for i, label in enumerate(italian_labels)}

# Сбор изображения для каждого класса
samples = {}
for img, label in train_loader:
    for i in range(len(label)):
        cls_idx = label[i].item()
        if cls_idx not in samples:
            samples[cls_idx] = img[i]
        if len(samples) == len(idx_to_italian):
            break
    if len(samples) == len(idx_to_italian):
        break

# Визуализация
n_classes = len(idx_to_italian)
plt.figure(figsize=(3 * n_classes, 3))

for idx in sorted(samples):
    img = samples[idx] * 0.5 + 0.5  # обратная нормализация
    np_img = img.numpy().transpose(1, 2, 0)

    label_it = idx_to_italian[idx]
    label_en = translate[label_it]

    plt.subplot(1, n_classes, idx + 1)
    plt.imshow(np_img)
    plt.axis('off')
    plt.title(label_en, fontsize=16)

plt.tight_layout()
plt.show()
```

## Создание модели

```python
class CustomCNN(nn.Module):
    def __init__(self, num_classes=10):
        super(CustomCNN, self).__init__()
        
        # Блок 1: Conv → BN → ReLU → Pool
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)  # 128→64
        
        # Блок 2: Conv → BN → ReLU → Pool
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)  # 64→32
        
        # Блок 3: Conv → BN → ReLU → Pool (заменяем на strided conv)
        self.conv3_1 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.bn3_1 = nn.BatchNorm2d(128)
        self.conv3_2 = nn.Conv2d(128, 128, kernel_size=3, stride=2, padding=1)  # 32→16
        self.bn3_2 = nn.BatchNorm2d(128)
        
        # Блок 4: Conv → BN → ReLU → Pool (заменяем на strided conv)
        self.conv4_1 = nn.Conv2d(128, 256, kernel_size=3, padding=1)
        self.bn4_1 = nn.BatchNorm2d(256)
        self.conv4_2 = nn.Conv2d(256, 256, kernel_size=3, stride=2, padding=1)  # 16→8
        

        self.fc1 = nn.Linear(256 * 8 * 8, 512)
        self.dropout1 = nn.Dropout(p=0.4)
        self.fc2 = nn.Linear(512, num_classes)

    def forward(self, x):
        # Блок 1
        x = self.pool1(F.relu(self.bn1(self.conv1(x))))
        
        # Блок 2
        x = self.pool2(F.relu(self.bn2(self.conv2(x))))
        
        # Блок 3
        x = F.relu(self.bn3_1(self.conv3_1(x)))
        x = F.relu(self.bn3_2(self.conv3_2(x)))  # Strided conv
        
        # Блок 4
        x = F.relu(self.bn4_1(self.conv4_1(x)))
        x = F.relu(self.conv4_2(x))  # Последний слой без BN
        
        # Flatten
        x = torch.flatten(x, 1)       # [B, 256*8*8]
        
        # Полносвязные слои с дропаутом между ними
        x = F.relu(self.fc1(x))
        x = self.dropout1(x)
        x = self.fc2(x)
        return x
```

```python
# Инициализируем модели, функции потерь и оптимизатора
model = CustomCNN(num_classes=10).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0005)

scheduler = ReduceLROnPlateau(
    optimizer,
    mode='max',
    factor=0.5,
    patience=5
)
```

```python
# Cтруктура модели
print("Model Summary...\n")
summary(model, input_size=(3, 128, 128))
```

### Обучение

```python
# колличество эпох
num_epochs = 25

train_losses = []
train_accuracies = []
val_losses = []
val_accuracies = []

best_val_accuracy = 0.0

model.to(device)
# Подсчет времени обучения
start_time = time.time()

for epoch in range(1, num_epochs + 1):
    # Обучение модели
    model.train()
    running_loss = 0.0
    correct_train = 0
    total_train = 0

    for images, labels in tqdm(train_loader,
                               desc=f"Эпоха {epoch}/{num_epochs} — обучение",
                               leave=True, ncols=80):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total_train += labels.size(0)
        correct_train += predicted.eq(labels).sum().item()

    epoch_train_loss = running_loss / len(train_loader)
    epoch_train_acc  = correct_train / total_train
    train_losses.append(epoch_train_loss)
    train_accuracies.append(epoch_train_acc)

    # Валидация модели
    model.eval()
    val_running_loss = 0.0
    correct_val = 0
    total_val = 0

    for images, labels in tqdm(val_loader,
                               desc=f"Эпоха {epoch}/{num_epochs} — валидация",
                               leave=True, ncols=80):
        images, labels = images.to(device), labels.to(device)

        with torch.no_grad():
            outputs = model(images)
            loss = criterion(outputs, labels)

        val_running_loss += loss.item()
        _, predicted = outputs.max(1)
        total_val += labels.size(0)
        correct_val += predicted.eq(labels).sum().item()

    epoch_val_loss = val_running_loss / len(val_loader)
    epoch_val_acc  = correct_val / total_val
    val_losses.append(epoch_val_loss)
    val_accuracies.append(epoch_val_acc)

    
    scheduler.step(epoch_val_acc)

    # Вывод результатов
    current_lr = optimizer.param_groups[0]['lr']
    print()  # пустая строка для удобства
    print(
        f"Эпоха [{epoch}/{num_epochs}] | "
        f"Train Loss: {epoch_train_loss:.4f}, "
        f"Train Acc: {epoch_train_acc*100:.2f}%, "
        f"Val Loss: {epoch_val_loss:.4f}, "
        f"Val Acc: {epoch_val_acc*100:.2f}%, "
        f"LR: {current_lr:.6f}"
    )

    # Сохранение лучшей модели
    if epoch_val_acc > best_val_accuracy:
        best_val_accuracy = epoch_val_acc
        torch.save(model.state_dict(), "best_model_optimize.pth")
        print("Лучшая модель сохранена")
    # Сохранение последней эпохи
    torch.save(model.state_dict(), "last_model_optimize.pth")

print("\nОбучение завершено.")
total_time = time.time() - start_time
print(f"Время обучения: {total_time/60:.2f} минут ({total_time:.1f} секунд)")
```

```python
#графики обучения
epochs = range(1, len(train_losses) + 1)

plt.figure(figsize=(10, 5))

# График потерь
plt.subplot(1, 2, 1)
plt.plot(epochs, train_losses, label='Train')
plt.plot(epochs, val_losses,   label='Validation')
plt.xlabel('Эпоха')
plt.ylabel('Потери')
plt.title('Потери к эпохе')
plt.legend()

# График точности
plt.subplot(1, 2, 2)
plt.plot(epochs, train_accuracies, label='Train')
plt.plot(epochs, val_accuracies,   label='Validation')
plt.xlabel('Эпоха')
plt.ylabel('точность')
plt.title('Точность к эпохе')
plt.legend()

plt.tight_layout()
plt.show()
```

# Проверка лучшей модели на тестовом наборе

```python
# Загрузка сохранённой лучшей модели
model.load_state_dict(torch.load('best_model_optimize.pth'))
model.to(device)
model.eval()

# Подсчёт потерь и точности на тесте
all_preds = []
all_labels = []
test_loss = 0.0
correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        labels = labels.to(device)

        outputs = model(images)
        loss = criterion(outputs, labels)
        test_loss += loss.item()

        _, preds = outputs.max(1)
        total += labels.size(0)
        correct += (preds == labels).sum().item()

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

final_test_loss = test_loss / len(test_loader)
final_test_acc  = correct / total

print(f"Тестовый результат (лучшая модель) — Loss: {final_test_loss:.4f} | Accuracy: {final_test_acc*100:.2f}%")
```

```python
# Получаем список уникальных имён классов
class_names = sorted(df_data["classes"].unique())

# Переводим на английский
class_names_en = [translate[c] for c in class_names]

# Отчёт по метрикам
print("Отчет по классификации:")
print(classification_report(all_labels, all_preds, target_names=class_names_en))

# Матрица ошибок
cm = confusion_matrix(all_labels, all_preds)

plt.figure(figsize=(10, 8))
sns.heatmap(
    cm, annot=True, fmt="d",
    xticklabels=class_names_en,
    yticklabels=class_names_en,
    cmap="Blues"
)
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.title("Матрица ошибок")
plt.tight_layout()
plt.show()
```

```python
# визуализаця предсказаний
num_images_to_show = 36
plt.figure(figsize=(14, 14))
count = 0

with torch.no_grad():
    for images, labels in test_loader:
        for i in range(images.size(0)):
            if count >= num_images_to_show:
                break

            image = images[i]
            label = labels[i]

            img_tensor = image.unsqueeze(0).to(device)
            output = model(img_tensor)
            _, pred_idx = output.max(1)

            img_disp = image.cpu() * 0.5 + 0.5
            img_disp = torch.clamp(img_disp, 0, 1).permute(1, 2, 0).numpy()

            label_it = class_names[label.item()]
            pred_it  = class_names[pred_idx.item()]

            classes_en = translate.get(label_it, label_it)
            pred_en    = translate.get(pred_it,  pred_it)

            title_color = "green" if pred_it == label_it else "red"
            title = f"Pred: {pred_en}\nTrue: {classes_en}"

            plt.subplot(6, 6, count + 1)
            plt.imshow(img_disp)
            plt.title(title, color=title_color, fontsize=8)
            plt.axis("off")

            count += 1
        if count >= num_images_to_show:
            break

plt.tight_layout()
plt.show()
```
